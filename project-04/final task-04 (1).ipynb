```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='Marketing_channel',
    y='Spend',
    estimator='sum'
)

plt.title("Total Marketing Spend by Channel", fontsize=14, fontweight='bold')
plt.xlabel("Marketing Channel")
plt.ylabel("Total Spend")
plt.xticks(rotation=45)

plt.show()
```

```python
"""Observation:
#PPC received the highest marketing spend among all channels, indicating that the company allocated a significant portion of its advertising budget to PPC campaigns.

#Business Insight:
#The organization appears to prioritize PPC as a major customer acquisition channel. However, higher spending should be evaluated alongside ROI to ensure efficient use of the marketing budget
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='Marketing_channel',
    y='Revenue',
    estimator='sum'
)

plt.title("Total Revenue by Marketing Channel", fontsize=14, fontweight='bold')
plt.xlabel("Marketing Channel")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)

plt.show()
```

```python
"""Observation:
#PPC generated the highest total revenue compared with the other marketing channels.

#Business Insight:
#PPC is a strong revenue-generating channel. If it also maintains a healthy ROI, it could justify additional investment.
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='Marketing_channel',
    y='ROI %',
    estimator='mean'
)

plt.title("Average ROI by Marketing Channel", fontsize=14, fontweight='bold')
plt.xlabel("Marketing Channel")
plt.ylabel("Average ROI (%)")
plt.xticks(rotation=45)

plt.show()
```

```python
"""Observation:
Referral marketing achieved the highest average ROI.

Business Insight:
Referral campaigns generate better returns per unit of investment, making them an efficient channel for future marketing efforts.
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Spend',
    y='Revenue',
    hue='Marketing_channel',
    s=100
)

plt.title("Spend vs Revenue")
plt.xlabel("Marketing Spend")
plt.ylabel("Revenue")
plt.show()
```

```python
"""#Observation:
#The scatter plot does not show a strong linear relationship between marketing spend and revenue.

#Business Insight:
#Increasing marketing spend alone does not guarantee higher revenue. Campaign strategy and targeting also influence performance.
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['ROI %'], bins=10, kde=True)

plt.title("Distribution of ROI")
plt.xlabel("ROI (%)")
plt.ylabel("Frequency")

plt.show()
```

```python
"""Observation:
Most campaigns achieved low to moderate ROI, while only a few campaigns generated exceptionally high returns.

Business Insight:
The company should analyze the high-performing campaigns to identify successful strategies that can be replicated.
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Spend',
    y='ROI %',
    hue='Marketing_channel',
    s=100
)

plt.title("ROI vs Marketing Spend")

plt.show()
```

```python
monthly = df.groupby(df['Date'].dt.to_period('M'))['Revenue'].sum()

plt.figure(figsize=(10,5))

monthly.plot(marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.grid(True)

plt.show()
```

```python
"""the november month generated highest revenue comparing to the all months"""
```

```python
plt.figure(figsize=(7,5))

corr = df[['Spend','Clicks','Revenue','ROI %']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Between Marketing Metrics")

plt.show()
```

```python
"""#Observation:
#A moderate negative correlation (-0.62) exists between Spend and ROI, while Revenue has a moderate positive relationship with ROI.

#Business Insight:
#Increasing marketing expenditure alone may reduce investment efficiency. Instead, the company should focus on optimizing campaign performance to improve ROI.#
```
